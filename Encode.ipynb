```python
#import
import re
import hashlib
import bchlib
import math
import os
import random
import numpy as np

#Read file into binary string
def readfile(file):
    file = open(file)
    data = file.read()
    file.close()
    result = ' '.join(map(bin, bytearray(data, 'utf-8')))
    #print("origin data:"+result)
    content_list = result.replace("0b",'',len(result))
    return content_list

def readfile_byte(file):
    file = open(file,'rb')
    data = file.read()
    file.close()
    return data

#Read file into binary string remove all whitespace and slice into two char  

def remove_space(string):
    output = "".join(string.split())
    return  output

def file_preprocess_1(string): 
    #input="".join(string.split()) 
    return re.findall('.',string)

def file_preprocess_2(string): 
    input="".join(string.split()) 
    return re.findall('.',string)

#binary to ATCG
def bintoDNA(input):
    output =[]
    DNA={"00":"A","01":"T","10":"C","11":"G"," ":" "}
        
    if(input[0]=='0'):
        output.append('C')
    else:
        output.append('G')
        
    #Huffman code 3bit to DNA    
    j=0
    for i in range(1,len(input)):
        if(output[j]=='A'):            
            if(input[i]=='0'):
                output.append('C')
            elif(input[i]=='1'):
                output.append('G')
            elif(input[i]==' '):
                output.append('T')
        elif(output[j]=='C'):            
            if(input[i]=='0'):
                output.append('G')
            elif(input[i]=='1'):
                output.append('T')
            elif(input[i]==' '):
                output.append('A')
        elif(output[j]=='G'):            
            if(input[i]=='0'):
                output.append('T')
            elif(input[i]=='1'):
                output.append('A')
            elif(input[i]==' '):
                output.append('C')
        elif(output[j]=='T'):            
            if(input[i]=='0'):
                output.append('A')
            elif(input[i]=='1'):
                output.append('C')
            elif(input[i]==' '):
                output.append('G')
        j=j+1
    return output  

def divide_chunks(input, chunk_size): 
    # looping till length l 
    for i in range(0, len(input), chunk_size):  
        yield input[i:i + chunk_size] 

def hex_to_DNA(input):
    output =[]
    DNA = {"0":"AA","1":"AT","2":"AC","3":"AG",
           "4":'TA',"5":"TT","6":"TC","7":"TG",
           "8":"CA","9":"CT","a":"CC","b":"CG",
           "c":"GA","d":"GT","e":"GC","f":"GG",}
    for i in input:
        output.append(DNA[i])
    return output


# perform conversion 
def str_to_int_list(test_list):
    for i in range(0, len(test_list)): 
        test_list[i] = int(test_list[i])
    return test_list


def int_to_str_list(list_string): 
    # Converting integer list to string list 
    # and joining the list using join() 
    res = "".join(map(str, list_string))
      
    return res 


def encode_file(filename):
    # create a bch object
    BCH_POLYNOMIAL = 8219
    BCH_BITS = 16
    bch = bchlib.BCH(BCH_POLYNOMIAL, BCH_BITS)
    
    #readfile
    data = bytearray(readfile_byte(filename))
    
    # encode and make a "packet"
    ecc = bch.encode(data)
    packet = data + ecc
    
    result = ' '.join(map(bin, packet))
    content_list = result.replace("0b",'',len(result))
    DNA =bintoDNA(file_preprocess_1(content_list)) 
    return "".join(DNA)

def generate_primer(filename):
    with open(filename, "rb") as f:
        buf = f.read()
    m = hashlib.md5(buf)
    input =  m.hexdigest()
    result = hex_to_DNA(input)
    return "".join(result)
    


print(encode_file ('DNASTORAGE.txt'))
print(len(generate_primer('DNASTORAGE.txt')))  
```
